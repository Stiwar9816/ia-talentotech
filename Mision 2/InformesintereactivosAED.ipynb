# Librería YData Profiling permite crear informes de AED con visualizaciones, estadística descriptivas, análisis de correlacción y detección de valores atípicos

```python
# Importar librerías necesarias
import pandas as pd
```

```python
# Cargar datos
df = pd.read_csv('Datos/calidad-del-aire-datos-historicos-diarios.csv',sep=";")
df.head()
```

```python
# Generar reporte automático
from ydata_profiling import ProfileReport

report = ProfileReport(df.sample(50000), title="Reporte de Análisis Exploratorio de Datos")
# Guardar el reporte
report_file= 'reporte_calidad_aire.html'
report.to_file(report_file)
report_file
```
