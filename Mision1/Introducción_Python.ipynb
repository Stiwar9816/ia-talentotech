## **Operaciones matem√°ticas y l√≥gicas**

Qu√© deben de saber:

*   **Operadores b√°sicos**: +, -, *, /, **, %, //
*   **Operadores de comparaci√≥n**: >, <, >=, <=, ==, !=
*   **Operadores condicionales**: *if* y *else*

````python
# Ejemplo operadores b√°sicos

n1=5
n2=10
sum = n1+n2
res = n2-n1
mul = n1*n2
div = n1/n2
mod = n2//n1
exp = n2**n1
print("Resultado suma:", sum)
print("Resultado resta:",res)
print("Resultado multiplicaci√≥n:",mul)
print("Resultado divisi√≥n:",div)
print("Resultado modulo:",mod)
print("Resultado exponencial:",exp)
````

````python
# Ejemplo operaciones de comparaci√≥n

# Variables
n1=10
n2=20

# (> - Mayor que)
comp = n1 > n2
print("Mayor que:",comp)

# (< - Menor que)
comp = n1 < n2
print("Menor que:",comp)

# (<= - Menor o igual)
comp = n1 <= n2
print("Menor o igual:",comp)

# (>= - Mayor o igual)
comp = n1 >= n2
print("Mayor o igual:",comp)

# (!= - Diferente)
comp = n1 != n2
print("Diferente:",comp)

# (== - Igual a)
comp = n1 == n2
print("Igual a:",comp)
````
